```python
#import libraries
import pandas as pd
import numpy as np
```

```python
#load dataset
df = pd.read_csv('/content/internet_users.csv')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
#Clean column names

df.columns = [
    'country',
    'rate_wb',
    'year_wb',
    'rate_itu',
    'year_itu',
    'users_cia',
    'year_cia',
    'notes'
]
```

```python
#Check for null values:
df.isnull().sum()
```

```python
#drop column
df = df.drop(columns=['notes'])
df.head()
```

```python
# Cleaning numerical columns

# Drop only if BOTH sources are missing
df = df.dropna(subset=['rate_wb', 'rate_itu'], how='all')

# Create average
df['avg_internet_rate'] = df[['rate_wb', 'rate_itu']].mean(axis=1)
```

```python
df.head()
```

```python
# Remove 'World' row
df = df[df['country'] != 'World']
```

```python
# Fix the type of year columns, from float to integer type
df['year_wb'] = df['year_wb'].astype('Int64')
df['year_itu'] = df['year_itu'].astype('Int64')
df['year_cia'] = df['year_cia'].astype('Int64')
```

```python
# Lightly clean cia_users column, and covert their type

df = df.dropna(subset=['users_cia'], how='all')
df['users_cia'] = df['users_cia'].astype('Int64')
```

```python
# Create a 'latest_year' column, as sometimes year of WB!= year of ITU

df['latest_year'] = df[['year_wb', 'year_itu']].max(axis=1)
df.head()
```

```python
# Save cleaned csv file
df.to_csv('internet_usage_cleaned.csv', index=False)
```
